```python
import os
import tmdbsimple as tmdb
import pandas as pd
import json
from tqdm import tqdm
from glob import glob
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
# filter only US movies
df = df[df.production_countries.str.contains("United States of America")]
df = df.sort_values(by='release_date', ascending=False)
df = df.reset_index(drop=True)
```

```python
df.columns
```

```python
target_year = 2017
# target_year = "" # use all new data
```

```python
US_movies_dict = {}
for fname in glob(f"dataset/*{target_year}_US_movies_dict.json"):
    with open(fname, "r") as f:
        US_movies_dict.update(json.load(f))
        print(len(US_movies_dict))
```

```python
records = []
for row in tqdm(US_movies_dict.values()):
    record = []
    for attr in df.columns:
        val = row[attr]
        if isinstance(val, list):
            val = json.dumps(val)
        record.append(val)
    records.append(record)
    
new_df = pd.DataFrame(records, columns=df.columns)
print(new_df.shape)
new_df.head()
```
